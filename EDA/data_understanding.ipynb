### Last Seasons Goals 

```python
new_seasons=fixtures_v2[(fixtures_v2["league.season"]>=2022) & (fixtures_v2["fixture.status.long"]=="Match Finished")]
new_seasons["Total_Goals"]=new_seasons["goals.home"] +new_seasons["goals.away"]
new_seasons
```

```python
def third_q(x):
    return x.quantile(0.75)
def first_q(x):
    return x.quantile(0.25)
```

```python
goal_sum= new_seasons.groupby("league.name",as_index=False)["Total_Goals"].aggregate(["min","max","mean",first_q,"median",third_q,"count"])
goal_sum.sort_values("median",ascending=False)
```

```python
only_valid["league.round"].unique()
```

```python
home= only_valid.drop(["teams.away.name"],axis=1).rename({"teams.home.name":"name"},axis=1)
away= only_valid.drop(["teams.home.name"],axis=1).rename({"teams.away.name":"name"},axis=1)
final_teams= pd.concat([home,away],axis=0)
final_teams
```

```python
grup_fix=final_teams.groupby(["name","league.id","league.season","league.country","league.name"],as_index=False)["league.round"].count()
grup_fix
```

```python
leagues=grup_fix.groupby(["league.id","league.season"],as_index=False)["league.round"].aggregate(["max","min"]).reset_index()
leagues
```

```python
#we will look leagues who has different teams with different match count

non_equal= leagues[leagues["max"]!=leagues["min"]]
non_equal
```

```python
non_equal["league.id"].unique()
```
